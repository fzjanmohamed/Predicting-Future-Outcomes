# **Model Analysis and Interpretation**

**Evaluating the model:**

Three main metrics used to evaluate a classfication model are accuracy, precision, and recall.

```python
plot_model(best_model, plot = 'feature')
```

**Feature Importance:**

The feature contributing most to the target of breast cancer recurrence according to this model is being in the premenopausal state.  A higher score means that the specific feature will have a larger effect on the model that is being used to predict breast cancer recurrence.  With this, you can understand relationship between features and target variable.  Being premenopausal, having the cancer in the left breast, and stage 3 malignancy are the three features contributing the most to the predictive power of the model.  Interestingly, the 40-49 years of age group is more important to the contribution of breast cancer recurrence than those in the 50-59 year old age group. 

```python
# install shap library
!pip install shap
```

```python
lightgbm = create_model('lightgbm')
```

```python
interpret_model(lightgbm)
```

SHAP value quantifies the contribution or the importance of each feature on the prediction of the entire model - it does not evaluate the quality of the prediction itself.  The feature that contributed the most to the prediction was stage 3 degree of malignancy.  High values of stage 3 degree of Malignancy have a high positive contribution to breast cancer recurrence and a low negative contribution  when values are low.  Tumour size (25-29 mm) contributed the least to the prediction.  High values of the number 0-2 invaded nodes had a high negative contribution, while low values of invaded nodes 0-2 had a high positive correlation with breast cancer recurrence.  This could mean that chances of recurrence are higher when more than 2 lymph nodes are invaded with the cancer.
